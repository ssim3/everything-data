## Measures of Central Tendency

Measures of Central Tendency,  are statistical values that provide information about the central values of a dataset. They are used to summarize and describe the distribution of data.

### Mean

The mean is the arithmetic average of a set of values. It is calculated by summing all the values in the dataset and dividing the sum by the total number of values. It is denoted by this formula:

<h1><center> $
\bar{x} = \frac{{\sum_{i=1}^{n} x_i}}{{n}}
$ </center></h1>

where:
- $\bar{x}$ represents the mean,
- $x_i$ represents each individual value in the dataset, and
- $n$ represents the total number of values in the dataset.


```python
# Getting the mean of Total Revenue by dividing sum / count

tr_sum = data['Total Revenue'].sum()
tr_count = data['Total Revenue'].count()

print("Total Revenue mean: ", tr_sum / tr_count)
```

```python
# Alternatively, using pandas
tr_mean = data["Total Revenue"].mean()

print("Total Revenue mean: ", tr_mean)
```

### Median

The median represents the middle value of a dataset. It is a way to find the "middle" value that divides the data into two equal halves. It is also known as the 50th percentile.


The formula for the median varies whether the dataset has an odd or even number of values:

<br>

- If $\mathrm{n}$ is odd, it is denoted by this formula:

<h1><center>$
\mathrm{Med}(x) = \frac{(n+1)}{2}
$</center></h1>

- If $\mathrm{n}$ is even, it is denoted by this formula:

<h1><center>$
\mathrm{Med}(x) =  \frac{(n / 2) + [(n / 2) + 1]} {2}
$</center></h1>

```python
# Getting the Median of Ad Revenue

# 1. Identify if Ad Revenue has even or odd number of data points
data['Ad Revenue'].count()

# 2. Since it has an odd number of data points, the formula for median is (n+1)/2
# (We have to sort the column and reset the index in order to index the correct value)
# (Moreover, we have to subtract one due to 0 indexes)
median_ar = data['Ad Revenue'].sort_values().reset_index(drop=True)[(data['Ad Revenue'].count() + 1) / 2 - 1]
print("Ad Revenue median: ", median_ar)
```

```python
# Alternatively, using pandas
data['Ad Revenue'].median()
```

### Mode

The mode represents the most frequently occurring value or values in a dataset. It is the value that appears with the highest frequency.

```python
# Getting the Mode of 'Ad gross Profit' column

# get the frequency of each unique value in the dataset, then index the first one (highest frequency)
mode_agp = data['Ad gross Profit'].value_counts().index[0]

print("Ad gross Profit mode: ", mode_agp)
```

```python
# Alternatively, using pandas
data['Ad gross Profit'].mode()[0]
```

## Measures of spread

### Variance and Standard Deviation

The Variance and Standard deviation are statistical measures used to quantify the spread or dispersion of data points in a dataset.



#### Variance

Variance measures the average squared deviation of each data point from the mean of the dataset. It gives us an idea of how much the data values vary or spread out from the average. 


- Higher variance -> larger spread of data points
- Lower variance -> more concentrated or clustered dataset.

The Population Variance is denoted by this formula:

<h1><center>
$
\sigma^2 = \frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2
$
</center></h1>

Where:
- $\sigma^2$ : variance
- $x_i$ : data point
- $\mu$ : mean

```python
# Finding variance of Ad Revenue with .var() function
print("Variance of Ad Revenue: ", data['Ad Revenue'].var())
```

#### Standard Deviation

Standard deviation is the square root of the variance. It is a widely used measure because it is expressed in the same unit as the data points, making it easier to interpret and compare. Standard deviation tells us how much, on average, each data point deviates from the mean.

- Higher sd -> larger spread of data points
- Lower sd -> datapoints are closer to the mean.

The Population Standard Deviation is denoted by this formula:

<h1><center>
$
\sigma = \sqrt{\sigma^2} = \sqrt{\frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2}
$
</center></h1>

- $\sigma$ : Standard Deviation
- $x_i$ : data point
- $\mu$ : mean

```python
# Finding variance of Ad Revenue with .var() function
print("Standard Deviation of Ad Revenue: ", data['Ad Revenue'].std())

# Ad Revenue has an SD of 107, indicating that the data points have a moderate spread.
```

### Quantiles and Quartiles

Quantiles are statistical measures that divide a dataset into equal-sized subsets. They help analyze the spread and distribution of values. The most commonly used quantiles are quartiles, quartiles divide data into four parts: Q1 (25%), Q2 (median), and Q3 (75%).

Quantiles are useful for understanding the spread of data and identifying outliers. They provide a way to describe the relative position of a particular value within a dataset and compare it to other values.

<div>

</div>

The .quantile() function in pandas can tell us the 0, 25th, 50th, 75th and 100th percentile of each column.
- (0th percentile -> minimum value)
- (100th percentile -> maximum value)

```python
# Q1
data['Total Revenue'].quantile(0.25)
```

```python
# Q2, a.k.a the median
data['Total Revenue'].quantile(0.5)
```

```python
# Q3
data['Total Revenue'].quantile(0.75)
```

The .describe() function tells us all the quartiles of every column in the dataset.

```python
# As we can see, Q3 of the Total Revenue column is 2501, while Q1 is 1449.

data.describe()
```

### Inter Quartile Range

The interquartile range (IQR) is a statistical measure that can be used to identify outliers in a dataset. It is calculated by subtracting the 3rd Quartile (Q3 or 75th %tile) by the 1st Quartile (Q1 or 25th %tile).

```python
plt.figure(figsize=(10, 6))

sns.boxplot(data=data, x='Gross Profit')

plt.title("Quartiles")

plt.show()
```

A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a visual summary of the key statistical measures and displays the spread and skewness of the data.

##### Box
- The central rectangular box represents the interquartile range (IQR). The lower boundary of the box represents the first quartile (Q1), and the upper boundary represents the third quartile (Q3).

##### Median
- The vertical line inside the rectangular box represent the median, or the second quartile (Q2).

##### Whiskers
- Whiskers, or the two lines at the end of each side of the figure, represents the range of the data. The lower whisker extends within 1.5 times the IQR below the first quartile. The upper whisker extends within 1.5 times the IQR above the third quartile. More on this later.

```python
# IQR of 'Gross Profit' column, 75th %tile - 25th %tile

gp_iqr = data['Gross Profit'].quantile(0.75) - data['Gross Profit'].quantile(0.25)
gp_iqr
```

### Outliers

Outliers are data points that significantly deviate from the overall pattern or distribution of a dataset. They are observations that are unusually far away from the majority of the data points and may indicate unusual behavior, errors, or anomalies in the data.

#### Identifying Outliers using IQR

To identify the outliers in a dataset using the IQR, we must define a <b>lower bound</b> and an <b>upper bound</b> for outliers using the following formulas:

Lower bound = Q1 - (1.5 * IQR)

Upper bound = Q3 + (1.5 * IQR)

(Multiplying the interquartile range (IQR) by 1.5 is a commonly used rule of thumb to determine the threshold for identifying outliers, hence the reason why the whiskers in the boxplot are 1.5 times the IQR below and above the first and third quartiles) 

```python
# Getting the lower and upper boundaries of Gross Profit

gp_lower = data['Gross Profit'].quantile(0.25) - (1.5 * gp_iqr)
gp_upper = data['Gross Profit'].quantile(0.75) + (1.5 * gp_iqr)

print(gp_lower, gp_upper)
```

To find the outliers in our dataset, we use np.where() to find the data points that lie below and above the lower and upper bound respectively.

```python
# Using np.where() to find outliers
outliers = np.where(data['Gross Profit'] > gp_upper, True, np.where(data['Gross Profit'] < gp_lower, True, False))

# The Gross Profit column does not have outliers
data.loc[outliers]
```
